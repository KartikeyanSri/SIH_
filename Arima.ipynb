```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Load the wheat data
df = pd.read_csv('wheatdata.csv')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce')

# Drop rows with missing 'Price' values
df = df.dropna(subset=['Price'])

# Get the unique 'Centre_Name' values
centres = df['Centre_Name'].unique()

# Define the inflation rate
inflation_rate = 0.03

def train_and_forecast(center_data, start_date, end_date, periods):
    """
    Trains ARIMA model and forecasts for the given number of periods.
    """
    center_data = center_data.set_index('Date')
    center_data = center_data['Price'].resample('M').mean().fillna(method='ffill')
    last_date = center_data.index[-1]
    train_data = center_data[:last_date]
    model = ARIMA(train_data, order=(5, 1, 0))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=periods)
    forecast_index = [last_date + timedelta(days=30*i) for i in range(1, periods+1)]
    return forecast, forecast_index

def adjust_for_inflation(prices, inflation_rate, periods):
    """
    Adjusts forecast prices for inflation over a given number of periods.
    """
    inflation_factor = (1 + inflation_rate / 12) ** np.arange(periods)
    return prices * inflation_factor

def forecast_between_dates(center_data, start_date, end_date, inflation_rate=0.03):
    """
    Forecasts wheat prices for a center between the given start and end dates.
    """
    # Get the last available date in the dataset
    last_date = center_data['Date'].max()

    # Ensure the start and end dates are later than the last available date
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    if start_date <= last_date or end_date <= start_date:
        raise ValueError("Start date must be later than the last available date, and end date must be after the start date.")

    # Calculate the number of months between the last date and the start date
    periods_before_start = (start_date.year - last_date.year) * 12 + (start_date.month - last_date.month)

    # Calculate the number of months between the start date and the end date
    total_periods = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)

    # Train and forecast for the total number of periods up to the end date
    forecast, forecast_index = train_and_forecast(center_data, start_date, end_date, total_periods + periods_before_start)

    # Adjust forecast for inflation
    forecast_adjusted = adjust_for_inflation(forecast, inflation_rate, total_periods + periods_before_start)

    # Create a DataFrame for the forecast
    forecast_df = pd.DataFrame({
        'Date': forecast_index,
        'Centre_Name': center_data['Centre_Name'].iloc[0],
        'Forecasted_Price': forecast_adjusted
    })

    # Filter the DataFrame to only include forecasts between the start and end dates
    forecast_df = forecast_df[(forecast_df['Date'] >= start_date) & (forecast_df['Date'] <= end_date)]

    return forecast_df

# Example: Forecast between two specific dates
start_date = '2025-01-01'
end_date =  '2026-12-01'

for center in centres:
    center_data = df[df['Centre_Name'] == center]

    try:
        forecast_df = forecast_between_dates(center_data, start_date, end_date, inflation_rate=0.03)

        # Show textual output of forecast
        print(f"Forecast for {center} between {start_date} and {end_date}:")
        print(forecast_df.to_string(index=False))  # Display all the forecast data as text without index

        # Plotting the graph for the forecasted prices
        plt.figure(figsize=(10, 6))
        plt.plot(forecast_df['Date'], forecast_df['Forecasted_Price'], color='red', linestyle='--', label='Forecasted Prices')
        plt.title(f'Wheat Price Forecast for {center} between {start_date} and {end_date}')
        plt.xlabel('Date')
        plt.ylabel('Price')
        plt.legend()
        plt.show()

    except ValueError as e:
        print(f"Error for {center}: {e}")
```
